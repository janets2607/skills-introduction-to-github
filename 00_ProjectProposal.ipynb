## Research Question

Does daily weather (such as temperature, sunlight, and precipitation) affect people's overall mood or sentiment as expressed on social media?



## Background and Prior Work

Weather is known to influence people in many ways and prior research in psychology has shown that factors like sunlight and temperature can affect mood and general well being. For example, more sunlight and comfortable temperatures are often linked with better mood, while gloomy or uncomfortable conditions can contribute to irritability and fatigue.<sup>1</sup> Recently, researchers have started using social media data to study these patterns since the language people use online can reflect how they feel. This has made it possible to examine weather and mood relationships across very large populations.

Large scale studies analyzing billions of posts across different cities found that temperature, precipitation, humidity, and cloud cover tend to be associated with more negative sentiment overall, even after removing posts that specifically talk about the weather.<sup>2</sup> While the effects on any single day are small, the patterns appear consistently across many locations and time periods. Some studies also report that extreme heat has a stronger negative effect on mood than moderate cold which matches other findings on how heat impacts stress and emotional regulation.<sup>2</sup>

Follow up research has expanded these results across different platforms. For example, a study combining Snapchat and Twitter data found that sentiment is generally more positive on sunnier days, and including weather, location, and time information improves mood prediction models.<sup>3</sup> Regional studies also show that extreme heat is linked with more negative or aggressive language, while colder temperatures tend to have smaller effects. This suggests that local climate and environmental context play a role in how people experience weather.<sup>4</sup>

Most studies on weather and mood rely on sentiment analysis which uses tools such as VADER or TextBlob to categorize social media posts as positive, negative, or neutral.<sup>5</sup> Researchers then calculate an average mood score for each day or city. Weather data typically comes from public sources such as Open Meteo API which provides daily metrics like temperature, rainfall, humidity, and sunlight.<sup>6</sup> Matching these daily weather conditions with the daily sentiment scores allows researchers to look for correlations such as whether warmer or sunnier days align with more positive language. These methods can't prove that weather directly causes mood shifts but they do highlight consistent patterns across large groups.

Our project takes a similar approach but on a smaller scale. We aim to examine whether day to day weather changes such as sunny vs. cloudy or warm vs. cold conditions are reflected in daily social media sentiment. To do this, we will collect weather data (temperature, rainfall, sunlight) and match it with daily sentiment data from a public social media dataset. Using basic data cleaning, visualization, and simple statistical analysis, we will explore whether the patterns found in larger studies also appear in our dataset. While we are not aiming to make causal claims, our goal is to see whether everyday weather conditions are associated with shifts in online mood expression.

## Hypothesis


We predict that people's online sentiment will become more positive on warmer, sunnier days and more negative on colder, rainy days.

In other words, when the weather is pleasant, featuring higher temperatures and more sunlight, social media posts will contain more happy/positive words and expressions. On days with low temperatures, heavy rainfall, or cloudy skies, posts will contain more negative words.

This prediction is based on previous research showing that sunlight exposure and comfortable weather can boost mood while gloomy or extreme conditions can lower it. We expect to see these emotional patterns reflected in the daily sentiment scores calculated from social media data.
